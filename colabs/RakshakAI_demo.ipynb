```python
from rakshakai.model import RakshakLightweightTransformer
from rakshakai.tokenizer import RakshakTokenizer
from rakshakai.inference import RakshakInference

engine = RakshakInference()
print(f'Model loaded: {engine.model.count_params()["total"]:,} params')
```

```python
snippets = [
    ('cursor.execute(f"SELECT * FROM users WHERE id = {user_id}")', 'python'),
    ('os.system("ping " + user_input)', 'python'),
    ('hashed = bcrypt.hashpw(password.encode(), bcrypt.gensalt())', 'python'),
    ('document.getElementById("out").innerHTML = userInput', 'javascript'),
    ('const API_KEY = "sk_live_abc123"', 'javascript'),
]

for code, lang in snippets:
    result = engine.predict(code, lang)
    icon = '🔴' if result['is_vulnerable'] else '✅'
    print(f"{icon} {result['label']:25s} ({result['confidence']:.3f}) | {code[:50]}")
```
